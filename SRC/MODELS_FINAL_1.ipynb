# **distilBERT MODEL**

````python
# Install dependencies (run once per session):
# !pip install torch transformers pandas scikit-learn matplotlib seaborn

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader, random_split
import pandas as pd
from transformers import (
    DistilBertTokenizerFast,
    DistilBertForSequenceClassification,
    get_linear_schedule_with_warmup
)
from torch.optim import AdamW
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# === 1. Ask user for Excel file path ===
EXCEL_FILE = input("📂 Enter the path to your Excel file (e.g., /content/FULL_DATA_UPDATE.xlsx): ")

# === 2. Global configuration ===
BATCH_SIZE = 8               # Option: try 16 or 32 if GPU memory allows
EPOCHS = 10                  # Increase the number of epochs
MAX_LEN = 512
LEARNING_RATE = 2e-5
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# === 3. Load and preprocess the data ===
df = pd.read_excel(EXCEL_FILE).dropna(subset=["Job Title", "Resume", "Seniority"])

# === 4. Create tokenizer and label mapping ===
tokenizer = DistilBertTokenizerFast.from_pretrained("distilbert-base-uncased")
label_names = sorted(df["Seniority"].unique())
label2id = {label: idx for idx, label in enumerate(label_names)}
id2label = {idx: label for label, idx in label2id.items()}

# === 5. Custom Dataset Class ===
class ResumeDataset(Dataset):
    def __init__(self, dataframe, tokenizer, max_len):
        self.tokenizer = tokenizer
        self.max_len = max_len

        self.inputs = []
        self.labels = []

        for _, row in dataframe.iterrows():
            text = f"Title: {row['Job Title']}\nResume: {row['Resume']}"
            encoding = self.tokenizer(
                text,
                truncation=True,
                padding="max_length",
                max_length=self.max_len,
                return_tensors="pt"
            )
            self.inputs.append({
                "input_ids": encoding["input_ids"].squeeze(0),
                "attention_mask": encoding["attention_mask"].squeeze(0)
            })
            self.labels.append(label2id[row["Seniority"]])

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        return (
            self.inputs[idx]["input_ids"],
            self.inputs[idx]["attention_mask"],
            self.labels[idx]
        )

# === 6. Split dataset into Training and Validation ===
full_dataset = ResumeDataset(df, tokenizer, MAX_LEN)
n_total = len(full_dataset)
n_train = int(0.8 * n_total)
n_val = n_total - n_train

train_ds, val_ds = random_split(full_dataset, [n_train, n_val])
train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True)
val_loader = DataLoader(val_ds, batch_size=BATCH_SIZE)

# === 7. Initialize DistilBERT model and optimizer ===
model = DistilBertForSequenceClassification.from_pretrained(
    "distilbert-base-uncased",
    num_labels=len(label2id)
).to(DEVICE)

optimizer = AdamW(model.parameters(), lr=LEARNING_RATE)
loss_fn = nn.CrossEntropyLoss()

# === 8. Set up Learning Rate Scheduler with warm-up ===
total_steps = EPOCHS * len(train_loader)
scheduler = get_linear_schedule_with_warmup(
    optimizer,
    num_warmup_steps=int(0.1 * total_steps),
    num_training_steps=total_steps
)

# === 9. Optional: Freeze lower layers for first few epochs ===
# Uncomment and adjust if desired:
# for name, param in model.distilbert.named_parameters():
#     if "transformer.layer.0" in name or "transformer.layer.1" in name:
#         param.requires_grad = False

# === 10. Mixed Precision (FP16) setup ===
use_amp = True if torch.cuda.is_available() else False
scaler = torch.cuda.amp.GradScaler() if use_amp else None

# === 11. Training Loop (no Early Stopping) ===
best_val_acc = 0.0

for epoch in range(1, EPOCHS + 1):
    model.train()
    total_train_loss = 0.0

    for input_ids, attention_mask, labels in train_loader:
        input_ids = input_ids.to(DEVICE)
        attention_mask = attention_mask.to(DEVICE)
        labels = labels.to(DEVICE)

        optimizer.zero_grad()

        if use_amp:
            with torch.cuda.amp.autocast():
                outputs = model(input_ids=input_ids, attention_mask=attention_mask)
                logits = outputs.logits
                loss = loss_fn(logits, labels)
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
        else:
            outputs = model(input_ids=input_ids, attention_mask=attention_mask)
            logits = outputs.logits
            loss = loss_fn(logits, labels)
            loss.backward()
            optimizer.step()

        scheduler.step()
        total_train_loss += loss.item()

    avg_train_loss = total_train_loss / len(train_loader)

    # === Validation Phase ===
    model.eval()
    val_loss_total = 0.0
    all_true = []
    all_pred = []

    with torch.no_grad():
        for input_ids, attention_mask, labels in val_loader:
            input_ids = input_ids.to(DEVICE)
            attention_mask = attention_mask.to(DEVICE)
            labels = labels.to(DEVICE)

            if use_amp:
                with torch.cuda.amp.autocast():
                    outputs = model(input_ids=input_ids, attention_mask=attention_mask)
                    logits = outputs.logits
                    loss = loss_fn(logits, labels)
            else:
                outputs = model(input_ids=input_ids, attention_mask=attention_mask)
                logits = outputs.logits
                loss = loss_fn(logits, labels)

            val_loss_total += loss.item()
            preds = torch.argmax(logits, dim=1).cpu().tolist()
            all_pred.extend(preds)
            all_true.extend(labels.cpu().tolist())

    avg_val_loss = val_loss_total / len(val_loader)
    val_acc = accuracy_score(all_true, all_pred) * 100

    print(
        f"Epoch {epoch}/{EPOCHS} | "
        f"Train Loss: {avg_train_loss:.4f} | "
        f"Val Loss: {avg_val_loss:.4f} | "
        f"Val Acc: {val_acc:.2f}%"
    )

    # === Save best model by Val Accuracy ===
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model.state_dict(), "best_distilbert_no_early.pt")

# === 12. Final Confusion Matrix ===
conf_mat = confusion_matrix(all_true, all_pred, labels=list(label2id.values()))
print(f"\n✅ Best DistilBERT Val Accuracy: {best_val_acc:.2f}%")
print("Final Confusion Matrix (Validation):")
print(conf_mat)

plt.figure(figsize=(6, 5))
sns.heatmap(
    conf_mat,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=label_names,
    yticklabels=label_names
)
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("DistilBERT – Confusion Matrix (Validation)")
plt.show()
````

**RoBERTa**

````python
# Install dependencies (run once per session):
# !pip install torch transformers pandas scikit-learn matplotlib seaborn

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader, random_split
import pandas as pd
from transformers import (
    RobertaTokenizerFast,
    RobertaForSequenceClassification,
    get_linear_schedule_with_warmup
)
from torch.optim import AdamW
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# === 1. Ask user for Excel file path (same file as for DistilBERT) ===
EXCEL_FILE = input("📂 Enter the path to your Excel file (e.g., /content/FULL_DATA_UPDATE.xlsx): ")

# === 2. Global configuration (same as DistilBERT for fair comparison) ===
BATCH_SIZE = 8               # Option: try 16 or 32
EPOCHS = 10                  # Increase to analyze effect
MAX_LEN = 512
LEARNING_RATE = 2e-5
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# === 3. Load and preprocess data once ===
df = pd.read_excel(EXCEL_FILE).dropna(subset=["Job Title", "Resume", "Seniority"])

label_names = sorted(df["Seniority"].unique())
label_map = {lab: idx for idx, lab in enumerate(label_names)}
id2label = {idx: lab for lab, idx in label_map.items()}

# === 4. Tokenizer for RoBERTa ===
roberta_tokenizer = RobertaTokenizerFast.from_pretrained("roberta-base")

# === 5. Custom Dataset for RoBERTa ===
class ClassifResumeDataset(Dataset):
    def __init__(self, dataframe, tokenizer, max_len):
        self.inputs = []
        self.labels = []
        self.max_len = max_len
        self.tokenizer = tokenizer

        for _, row in dataframe.iterrows():
            text = f"Title: {row['Job Title']}\nResume: {row['Resume']}"
            encoding = self.tokenizer(
                text,
                truncation=True,
                padding="max_length",
                max_length=self.max_len,
                return_tensors="pt"
            )
            self.inputs.append({
                "input_ids": encoding["input_ids"].squeeze(0),
                "attention_mask": encoding["attention_mask"].squeeze(0)
            })
            self.labels.append(label_map[row["Seniority"]])

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        return (
            self.inputs[idx]["input_ids"],
            self.inputs[idx]["attention_mask"],
            self.labels[idx]
        )

# === 6. Split dataset and create DataLoaders ===
full_dataset = ClassifResumeDataset(df, roberta_tokenizer, MAX_LEN)
n_total = len(full_dataset)
n_train = int(0.8 * n_total)
n_val = n_total - n_train

train_ds, val_ds = random_split(full_dataset, [n_train, n_val])
train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True)
val_loader = DataLoader(val_ds, batch_size=BATCH_SIZE)

# === 7. Initialize RoBERTa model ===
model_rb = RobertaForSequenceClassification.from_pretrained(
    "roberta-base",
    num_labels=len(label_map)
).to(DEVICE)

optimizer_rb = AdamW(model_rb.parameters(), lr=LEARNING_RATE)
criterion_rb = nn.CrossEntropyLoss()

# === 8. Scheduler setup ===
total_steps = EPOCHS * len(train_loader)
scheduler_rb = get_linear_schedule_with_warmup(
    optimizer_rb,
    num_warmup_steps=int(0.1 * total_steps),
    num_training_steps=total_steps
)

# === 9. Optional: Freeze lower layers ===
# for name, param in model_rb.roberta.named_parameters():
#     if "encoder.layer.0" in name or "encoder.layer.1" in name:
#         param.requires_grad = False

# === 10. Mixed Precision (FP16) setup ===
use_amp = True if torch.cuda.is_available() else False
scaler_rb = torch.cuda.amp.GradScaler() if use_amp else None

# === 11. Training + Validation Loop (no Early Stopping) ===
best_val_acc = 0.0

for epoch in range(1, EPOCHS + 1):
    # — Training Phase —
    model_rb.train()
    total_train_loss = 0.0

    for input_ids, attention_mask, labels in train_loader:
        input_ids = input_ids.to(DEVICE)
        attention_mask = attention_mask.to(DEVICE)
        labels = torch.tensor(labels).to(DEVICE)

        optimizer_rb.zero_grad()

        if use_amp:
            with torch.cuda.amp.autocast():
                outputs = model_rb(input_ids=input_ids, attention_mask=attention_mask)
                logits = outputs.logits
                loss = criterion_rb(logits, labels)
            scaler_rb.scale(loss).backward()
            scaler_rb.step(optimizer_rb)
            scaler_rb.update()
        else:
            outputs = model_rb(input_ids=input_ids, attention_mask=attention_mask)
            logits = outputs.logits
            loss = criterion_rb(logits, labels)
            loss.backward()
            optimizer_rb.step()

        scheduler_rb.step()
        total_train_loss += loss.item()

    avg_train_loss = total_train_loss / len(train_loader)

    # — Validation Phase —
    model_rb.eval()
    total_val_loss = 0.0
    all_true = []
    all_pred = []

    with torch.no_grad():
        for input_ids, attention_mask, labels in val_loader:
            input_ids = input_ids.to(DEVICE)
            attention_mask = attention_mask.to(DEVICE)
            labels = torch.tensor(labels).to(DEVICE)

            if use_amp:
                with torch.cuda.amp.autocast():
                    outputs = model_rb(input_ids=input_ids, attention_mask=attention_mask)
                    logits = outputs.logits
                    loss = criterion_rb(logits, labels)
            else:
                outputs = model_rb(input_ids=input_ids, attention_mask=attention_mask)
                logits = outputs.logits
                loss = criterion_rb(logits, labels)

            total_val_loss += loss.item()
            preds = torch.argmax(logits, dim=1).cpu().tolist()
            all_pred.extend(preds)
            all_true.extend(labels.cpu().tolist())

    avg_val_loss = total_val_loss / len(val_loader)
    val_acc = accuracy_score(all_true, all_pred) * 100
    conf_mat = confusion_matrix(all_true, all_pred, labels=list(range(len(label_map))))

    print(
        f"Epoch {epoch}/{EPOCHS} | "
        f"Train Loss: {avg_train_loss:.4f} | "
        f"Val Loss: {avg_val_loss:.4f} | "
        f"Val Acc: {val_acc:.2f}%"
    )

    # Save best checkpoint by Val Accuracy
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model_rb.state_dict(), "best_roberta_no_early.pt")

# === 12. Final Confusion Matrix ===
print(f"\n✅ Best RoBERTa Val Accuracy: {best_val_acc:.2f}%")
print("Final Confusion Matrix (Validation):")
print(conf_mat)

plt.figure(figsize=(6, 5))
sns.heatmap(
    conf_mat,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=label_names,
    yticklabels=label_names
)
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("RoBERTa – Confusion Matrix (Validation)")
plt.show()
````

# **GPT MODEL**

````python
# =============================
# Resume Seniority Classification with OpenRouter GPT API (with 2-key fallback)
# =============================

# 1. Install required libraries (if needed)
# !pip install pandas requests scikit-learn matplotlib seaborn

import requests
import json
import pandas as pd
import time
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Configuration
API_KEYS = [
   
]
MODEL = "openai/gpt-4.1"
API_URL = "https://openrouter.ai/api/v1/chat/completions"

# 3. Load shared data and normalize Seniority to lowercase
df = (
    pd.read_excel("FULL_DATA_UPDATE.xlsx")
      .dropna(subset=["Resume", "Job Title", "Seniority"])
)
df["Seniority"] = df["Seniority"].astype(str).str.strip().str.lower()

# 4. GPT Classification Function with 2-key fallback
def classify_resume(title, resume_text):
    prompt = f"""
You are an expert HR analyst.
Classify the seniority level of the resume below as 'junior', 'mid', or 'senior' based only on the job title and resume content.
Return only this JSON format:
{{"level": "<junior|mid|senior>"}}

Job Title: {title}
Resume:
```{resume_text[:12000]}```
""".strip()

    for key in API_KEYS.copy():  # iterate over a copy so we can modify API_KEYS if needed
        headers = {
            "Authorization": f"Bearer {key}",
            "Content-Type": "application/json"
        }
        data = {
            "model": MODEL,
            "messages": [
                {"role": "system", "content": "Only return valid JSON."},
                {"role": "user", "content": prompt}
            ],
            "temperature": 0,
            "max_tokens": 50
        }

        try:
            response = requests.post(API_URL, headers=headers, json=data, timeout=30)
            response.raise_for_status()
            content = response.json().get("choices", [{}])[0].get("message", {}).get("content", "")
            result = json.loads(content)
            level = result.get("level", "").strip().lower()
            if level in {"junior", "mid", "senior"}:
                return level
            else:
                # JSON is valid but the output is not one of the expected labels
                print(f"⚠️ Unexpected JSON format with key {key[:10]}...: {content}")
                continue

        except requests.HTTPError as http_err:
            status = getattr(http_err.response, "status_code", None)
            if status == 402:
                # If we got 402, it means the key has expired/been blocked
                print(f"⚠️ Key {key[:10]}... blocked (402 Payment Required). Removing from rotation.")
                API_KEYS.remove(key)
                continue
            else:
                print(f"⚠️ HTTP error with key {key[:10]}...: {http_err}")
                continue

        except (requests.RequestException, json.JSONDecodeError) as e:
            # Communication or JSON decode error – try the next key
            print(f"⚠️ Error with key {key[:10]}...: {e}")
            continue

    # If we reached here, all keys have failed
    return "error"

# 5. Apply GPT classification
predicted = []
for i, row in df.iterrows():
    print(f"🔍 Classifying row {i+1}/{len(df)}: {row['Job Title']}")
    level = classify_resume(row["Job Title"], row["Resume"])
    print(f"→ Predicted: {level} | Actual: {row['Seniority']}")
    predicted.append(level)
    time.sleep(1.5)  # Respect rate limits

# 6. Evaluation
df["Predicted"] = predicted
filtered_df = df[df["Predicted"] != "error"].copy()
y_true = filtered_df["Seniority"]
y_pred = filtered_df["Predicted"]

# 6.1 Classification Report and Accuracy
labels = ["junior", "mid", "senior"]
print("\n📊 GPT Classification Report:")
print(classification_report(y_true, y_pred, labels=labels, digits=4))
accuracy = accuracy_score(y_true, y_pred)
print(f"✅ Accuracy: {accuracy:.4f}")

# 7. Confusion Matrix
conf_matrix = confusion_matrix(y_true, y_pred, labels=labels)
plt.figure(figsize=(6, 5))
sns.heatmap(
    conf_matrix,
    annot=True,
    fmt="d",
    cmap="Purples",
    xticklabels=labels,
    yticklabels=labels
)
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix - GPT Classification")
plt.show()

# 8. Save results
filtered_df.to_csv("classified_resumes_gpt.csv", index=False)
print("✅ File saved: classified_resumes_gpt.csv")
````
